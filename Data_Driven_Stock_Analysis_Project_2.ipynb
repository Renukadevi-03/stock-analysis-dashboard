```python
import os
import yaml
import pandas as pd

all_data = []

base_path = '/content/data/data'

for root, dirs, files in os.walk(base_path):

    for file in files:

        if file.endswith('.yaml'):

            filepath = os.path.join(root, file)

            with open(filepath, 'r') as f:

                data = yaml.safe_load(f)

                # data is a LIST
                for item in data:

                    row = {
                        "Ticker": item.get("Ticker"),
                        "Date": item.get("date"),
                        "Month": item.get("month"),
                        "Open": item.get("open"),
                        "High": item.get("high"),
                        "Low": item.get("low"),
                        "Close": item.get("close"),
                        "Volume": item.get("volume")
                    }

                    all_data.append(row)

df = pd.DataFrame(all_data)

print(df.head())
print(df.shape)
```

Calculate Yearly Return

```python
returns = []

for ticker in df["Ticker"].unique():

    stock_df = df[df["Ticker"] == ticker]

    first_close = stock_df.iloc[0]["Close"]

    last_close = stock_df.iloc[-1]["Close"]

    yearly_return = (
        (last_close - first_close)
        / first_close
    ) * 100

    returns.append([
        ticker,
        yearly_return
    ])

returns_df = pd.DataFrame(
    returns,
    columns=["Ticker", "Yearly_Return"]
)

print(returns_df.head())
```

Top 10 Green Stocks

```python
top_green = returns_df.sort_values(
    by="Yearly_Return",
    ascending=False
).head(10)

print(top_green)
```

Top 10 Red Stocks

```python
top_red = returns_df.sort_values(
    by="Yearly_Return"
).head(10)

print(top_red)
```

Market Summary

```python
green_count = (
    returns_df["Yearly_Return"] > 0
).sum()

red_count = (
    returns_df["Yearly_Return"] < 0
).sum()

average_price = df["Close"].mean()

average_volume = df["Volume"].mean()

print("Green Stocks :", green_count)

print("Red Stocks :", red_count)

print("Average Price :", round(average_price, 2))

print("Average Volume :", round(average_volume, 2))
```

Daily Return

```python
df["Daily_Return"] = (
    df.groupby("Ticker")["Close"]
    .pct_change()
)

print(df.head())
```

Volatility Calculation

```python
volatility_df = (
    df.groupby("Ticker")["Daily_Return"]
    .std()
    .reset_index()
)

volatility_df.rename(
    columns={
        "Daily_Return": "Volatility"
    },
    inplace=True
)

top_volatile = volatility_df.sort_values(
    by="Volatility",
    ascending=False
).head(10)

print(top_volatile)
```

Volatility Bar Chart

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

plt.bar(
    top_volatile["Ticker"],
    top_volatile["Volatility"]
)

plt.title("Top 10 Most Volatile Stocks")

plt.xlabel("Ticker")

plt.ylabel("Volatility")

plt.xticks(rotation=45)

plt.show()
```

Sector-wise Performance Analysis

Merge Sector Data With Returns

```python
sector_returns = pd.merge(
    returns_df,
    sector_df,
    left_on="Ticker",
    right_on="Symbol",
    how="left"
)

print(sector_returns.head())
```

Average Return By Sector

```python
sector_performance = (
    sector_returns.groupby("sector")[
        "Yearly_Return"
    ]
    .mean()
    .reset_index()
)

print(sector_performance)
```

```python
sector_df["Symbol"] = (
    sector_df["Symbol"]
    .str.split(":")
    .str[-1]
    .str.strip()
    .str.upper()
)

print(sector_df["Symbol"].head())
```

```python
sector_returns = pd.merge(
    returns_df,
    sector_df,
    left_on="Ticker",
    right_on="Symbol",
    how="inner"
)

print(sector_returns.head())

print(sector_returns.shape)
```

Sector Performance

```python
sector_performance = (
    sector_returns.groupby("sector")[
        "Yearly_Return"
    ]
    .mean()
    .reset_index()
)

print(sector_performance)
```

Sector Bar Chart

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

plt.bar(
    sector_performance["sector"],
    sector_performance["Yearly_Return"]
)

plt.title("Average Yearly Return by Sector")

plt.xlabel("Sector")

plt.ylabel("Average Yearly Return")

plt.xticks(rotation=45)

plt.show()
```

Calculate Cumulative Return

```python
# Calculate cumulative return

df["Cumulative_Return"] = (
    1 + df["Daily_Return"]
)

df["Cumulative_Return"] = (
    df.groupby("Ticker")["Cumulative_Return"]
    .cumprod() - 1
)

print(df.head())
```

```python
top5 = returns_df.sort_values(
    by="Yearly_Return",
    ascending=False
).head(5)

top5_tickers = top5["Ticker"].tolist()

print(top5_tickers)
```

Line Chart

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,7))

for ticker in top5_tickers:

    stock_df = df[df["Ticker"] == ticker]

    plt.plot(
        stock_df["Date"],
        stock_df["Cumulative_Return"],
        label=ticker
    )

plt.title("Cumulative Return of Top 5 Stocks")

plt.xlabel("Date")

plt.ylabel("Cumulative Return")

plt.legend()

plt.grid(True)

plt.show()
```

Monthly Top 5 Gainers & Losers

Create Month Column

```python
df["Month"] = df["Date"].dt.to_period("M")

print(df.head())
```

Monthly Return Calculation

```python
monthly_returns = []

for (month, ticker), group in df.groupby(["Month", "Ticker"]):

    group = group.sort_values("Date")

    first_close = group.iloc[0]["Close"]

    last_close = group.iloc[-1]["Close"]

    monthly_return = (
        (last_close - first_close)
        / first_close
    ) * 100

    monthly_returns.append([
        month,
        ticker,
        monthly_return
    ])

monthly_returns_df = pd.DataFrame(
    monthly_returns,
    columns=[
        "Month",
        "Ticker",
        "Monthly_Return"
    ]
)

print(monthly_returns_df.head())
```

Top 5 Gainers & Losers Each Month

```python
for month in monthly_returns_df["Month"].unique():

    month_data = monthly_returns_df[
        monthly_returns_df["Month"] == month
    ]

    top_gainers = month_data.sort_values(
        by="Monthly_Return",
        ascending=False
    ).head(5)

    top_losers = month_data.sort_values(
        by="Monthly_Return"
    ).head(5)

    print("\nMonth :", month)

    print("\nTop Gainers")
    print(top_gainers)

    print("\nTop Losers")
    print(top_losers)
```

Create Monthly Charts

```python
import matplotlib.pyplot as plt

months = monthly_returns_df["Month"].unique()

for month in months:

    month_data = monthly_returns_df[
        monthly_returns_df["Month"] == month
    ]

    top_gainers = month_data.sort_values(
        by="Monthly_Return",
        ascending=False
    ).head(5)

    top_losers = month_data.sort_values(
        by="Monthly_Return"
    ).head(5)

    # Create figure
    plt.figure(figsize=(12,5))

    # Gainers chart
    plt.subplot(1,2,1)

    plt.bar(
        top_gainers["Ticker"],
        top_gainers["Monthly_Return"]
    )

    plt.title(f"Top 5 Gainers - {month}")

    plt.xticks(rotation=45)

    plt.ylabel("Monthly Return")

    # Losers chart
    plt.subplot(1,2,2)

    plt.bar(
        top_losers["Ticker"],
        top_losers["Monthly_Return"]
    )

    plt.title(f"Top 5 Losers - {month}")

    plt.xticks(rotation=45)

    plt.ylabel("Monthly Return")

    plt.tight_layout()

    plt.show()
```
